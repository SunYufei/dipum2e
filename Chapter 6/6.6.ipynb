## 6.6.2

```matlab
f = imread('Fig0630(a).tif');
mask = roipoly(f);

red = immultiply(mask, f(:, :, 1));
green = immultiply(mask, f(:, :, 2));
blue = immultiply(mask, f(:, :, 3));

g = cat(3, red, green, blue);
imshow(g);
```
