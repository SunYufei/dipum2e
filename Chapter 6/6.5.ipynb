```matlab
fc = imread('Fig0622(a).tif');

h = rgb2hsi(fc);

subplot(221), imshow(fc)
subplot(222), imshow(fc(:, :, 1));
subplot(223), imshow(fc(:, :, 2));
subplot(224), imshow(fc(:, :, 3));
```

```matlab
subplot(221), imshow(h(:, :, 1));
subplot(222), imshow(h(:, :, 2));
subplot(223), imshow(h(:, :, 3));
```

```matlab
H = h(:, :, 1);
S = h(:, :, 2);
I = h(:, :, 3);

w = fspecial('average', 25);

H_filtered = imfilter(H, w, 'replicate');
I_filtered = imfilter(I, w, 'replicate');
S_filtered = imfilter(S, w, 'replicate');

h1 = cat(3, H_filtered, S, I);
h2 = cat(3, H, S, I_filtered);
h3 = cat(3, H, S_filtered, I);

f1 = hsi2rgb(h1);
imshow(f1)
```

```matlab
f2 = hsi2rgb(h2);
imshow(f2)
```

```matlab
f3 = hsi2rgb(h3);
imshow(f3)
```
