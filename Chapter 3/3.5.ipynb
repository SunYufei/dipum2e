# 3.5 在频域中直接生成滤波器
## 3.5.1 建立网络数组以实现频域滤波器

```matlab
[U, V] = dftuv(8, 5);
DSQ = U .^ 2 + V .^ 2
```

```matlab
fftshift(DSQ)
```

## 3.5.2 频域低通(平滑)滤波器

```matlab
f = imread('Fig0313(a).tif');
[f, revertclass] = tofloat(f);
PQ = paddedsize(size(f));
[U, V] = dftuv(PQ(1), PQ(2));
D = hypot(U, V);
D0 = 0.05 * PQ(2);
F = fft2(f, PQ(1), PQ(2));
H = exp(-(D.^2) / (2 * (D0^2)));
g = dftfilt(f, H);
g = revertclass(g);

subplot(221), imshow(f, []);
subplot(222), imshow(fftshift(H));
subplot(223), imshow(log(1 + abs(fftshift(F))), []);
subplot(224), imshow(g);
```

## 3.5.3 线框及表面绘制

```matlab
H = fftshift(lpfilter('gaussian', 500, 500, 50));
mesh(double(H(1:10:500, 1:10:500)));
axis tight
```

```matlab
H = fftshift(lpfilter('gaussian', 500, 500, 50));
mesh(double(H(1:10:500, 1:10:500)));
axis tight
colormap([0 0 0])
axis off
grid off
```

```matlab
H = fftshift(lpfilter('gaussian', 500, 500, 50));
mesh(double(H(1:10:500, 1:10:500)));
axis tight
colormap([0 0 0])
axis off
grid off
view(-25, 30);
```

```matlab
H = fftshift(lpfilter('gaussian', 500, 500, 50));
mesh(double(H(1:10:500, 1:10:500)));
axis tight
colormap([0 0 0])
axis off
grid off
view(-25, 0);
```
