## 3.7.2 陷波带阻和陷波带通滤波器

```matlab
f = imread('Fig0321(a).tif');
subplot(231), imshow(f);

[M N] = size(f);
[f, revertclass] = tofloat(f);
F = fft2(f);
S = gscale(log(1 + abs(fftshift(F))));
subplot(232), imshow(S);

C1 = [99 154; 128 163];
H1 = cnotch('gaussian', 'reject', M, N, C1, 5);
P1 = gscale(fftshift(H1) .* (tofloat(S)));
subplot(233), imshow(P1);

g1 = dftfilt(f, H1);
g1 = revertclass(g1);
subplot(234), imshow(g1);

C2 = [99 154; 128 163; 49 160; 133 233; 55 132; 108 225; 112 74];
H2 = cnotch('gaussian', 'reject', M, N, C2, 5);
P2 = gscale(fftshift(H2) .* (tofloat(S)));
subplot(235), imshow(P2);

g2 = dftfilt(f, H2);
g2 = revertclass(g2);
subplot(236), imshow(g2)
```

```matlab
f = imread('Fig0322(a).tif');
subplot(221), imshow(f);

[M, N] = size(f);
[f, revertclass] = tofloat(f);
F = fft2(f);
S = gscale(log(1 + abs(fftshift(F))));
subplot(222), imshow(S);

H = recnotch('reject', 'vertical', M, N, 3, 15, 15);
subplot(223), imshow(fftshift(H));

g = dftfilt(f, H);
g = revertclass(g);
subplot(224), imshow(g);
```

```matlab
f = imread('Fig0323(a).tif');
subplot(131), imshow(f);

Hrecpass = recnotch('pass', 'vertical', M, N, 3, 15, 15);
interference = dftfilt(f, Hrecpass);

subplot(132), imshow(fftshift(Hrecpass));

interference = gscale(interference);
subplot(133), imshow(interference);
```
