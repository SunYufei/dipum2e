# 3.4 从空域滤波器获得频域滤波器

```matlab
f = imread('Fig0309(a).tif');
f = tofloat(f);
F = fft2(f);
S = fftshift(log(1 + abs(F)));
subplot(121), imshow(f);
subplot(122), imshow(S, []);
```

```matlab
h = fspecial('sobel')'
freqz2(h)
```

```matlab
f = imread('Fig0309(a).tif');

PQ = paddedsize(size(f));
H = freqz2(h, PQ(1), PQ(2));
H1 = ifftshift(H)

subplot(121), imshow(abs(H), []);
subplot(122), imshow(abs(H1), []);
```

```matlab
f = imread('Fig0309(a).tif');
h = fspecial('sobel')';

PQ = paddedsize(size(f));
H = freqz2(h, PQ(1), PQ(2));
H1 = ifftshift(H);

gs = imfilter(f, h);
gf = dftfilt(f, H1);
subplot(221), imshow(gs, []);
subplot(222), imshow(gf, []);
subplot(223), imshow(abs(gs), []);
subplot(224), imshow(abs(gf), []);
```

```matlab
subplot(121), imshow(abs(gs) > 0.2 * abs(max(gs(:))));
subplot(122), imshow(abs(gf) > 0.2 * abs(max(gf(:))));
```
