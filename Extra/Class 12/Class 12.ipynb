### Example 8.13: Block transform coding with the DFT, WHT, and DCT

DFT: discrete Fourier transform

WHT: Walsh-Hadamard transform

DCT: discrete cosine transform

```matlab
f = imread('Fig0809(a).tif');
f = im2double(f);
imshow(f);
```

Approximations of Fig. 8.9(a) using the Fourier transform.

```matlab
dft = fft2(f);
dft = ifft2(dft);
figure, imshow(dft);
```

The corresponding scaled error image.

```matlab
figure, imshow(f - dft, []);
```

Approximations of Fig. 8.9(a) using the Walsh-Hadamard transform.

```matlab
wht = fwht(f);
wht = ifwht(wht);
figure, imshow(wht);
```

The corresponding scaled error image.

```matlab
figure, imshow(f - wht, []);
```

Approximations of Fig. 8.9(a) using the cosine transform.

```matlab
dct = dct2(f);
dct = idct2(dct);
figure, imshow(dct);
```

The corresponding scaled error image.

```matlab
figure, imshow(f - dct, []);
```

### JPEG

One of the mose popular and comprehensive continuous tone, still frame compression standards is the JPEG standard. It defines three different coding systems: (1) a lossy baseline coding system, which is based on the DCT and is adequate for mose compression applications; (2) an extended coding system for greater compression, higher precision, or progressive reconstruction applications; and (3) a lossless independent coding system for reversible compression.

### Example 8.18: Illustration of JPEG coding.

```matlab
f = imread('Fig0809(a).tif');
imshow(f);
```

The next images show two JPEG approximations of the monochrome image in Fig. 8.9(a) with two different compressions and the differences between the original image and the reconstructed images.

```matlab
g = im2jpeg(f, 0.25);
g = jpeg2im(g);
figure, imshow(g);
```

```matlab
figure, imshow(f - g, []);
```

```matlab
h = im2jpeg(f, 0.52);
h = jpeg2im(h);
figure, imshow(h);
```

```matlab
figure, imshow(f - h, []);
```
