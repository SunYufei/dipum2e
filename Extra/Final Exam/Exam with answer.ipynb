## 2. Draw the following function in one figure shown below.

$y=5\pi(\sin(x)+\cos(x))+x$, where, $-5\pi\leq x\leq 5\pi$

If you get a plot which is not smooth like this, you will get 8 score.

```matlab
x = -5 * pi : 5 * pi;
y = 5 * pi * (sin(x) + cos(x)) + x;
plot(x, y)
title('[Type your real name here]')
xlabel('x')
ylabel('y')
legend('y = f(x)')
```
