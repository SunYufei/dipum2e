# 4.3 仅有噪声的复原——空间滤波

## 4.3.1 空间噪声滤波器

```matlab
f = imread('Fig0405(a).tif');
[M, N] = size(f);

R = imnoise2('salt & pepper', M, N, 0.1, 0);
gp = f;
gp(R == 0) = 0;

subplot(231), imshow(gp, []);

R2 = imnoise2('salt & pepper', M, N, 0, 0.1);
gs = f;
gs(R2 == 1) = 255;

subplot(232), imshow(gs, []);

fp = spfilt(gp, 'chmean', 3, 3, 1.5);
subplot(233), imshow(fp, []);

fs = spfilt(gs, 'chmean', 3, 3, -1.5);
subplot(234), imshow(fs, []);

fpmax = spfilt(gp, 'max', 3, 3);
subplot(235), imshow(fpmax, []);

fsmin = spfilt(gs, 'min', 3, 3);
subplot(236), imshow(fsmin, []);
```

## 4.3.2 自适应空间滤波器

```matlab
f = imread('Fig0406(a).tif');
subplot(221), imshow(f);

g = imnoise(f, 'salt & pepper', 0.25);
subplot(222), imshow(g);

f1 = medfilt2(g, [7 7], 'symmetric');
subplot(223), imshow(f1);

f2 = adpmedian(g, 7);
subplot(224), imshow(f2);
```
