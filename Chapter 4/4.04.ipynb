# 4.5 退化函数建模

```matlab
f = checkerboard(8);
imshow(pixeldup(f, 8), []);
```

```matlab
PSF = fspecial('motion', 7, 45)
gb = imfilter(f, PSF, 'circular');
imshow(pixeldup(gb, 8), []);
```

```matlab
noise = imnoise2('Gaussian', size(f, 1), size(f, 2), 0, sqrt(0.001));
imshow(pixeldup(noise, 8), []);
```

```matlab
g = gb + noise;
imshow(pixeldup(g, 8), []);
```
