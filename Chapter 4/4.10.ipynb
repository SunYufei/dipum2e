# 4.10 盲去卷积

```matlab
PSF = fspecial('gaussian', 7, 10);
imshow(pixeldup(PSF, 73), []);
```

```matlab
g = checkerboard(8);
PSF = fspecial('gaussian', 7, 10);

SD = 0.01;
g = imnoise(imfilter(g, PSF), 'gaussian', 0, SD ^ 2);

INITPSF = ones(size(PSF));
LIM = ceil(size(PSF, 1) / 2);
NUMIT = 5;
DAMPAR = 10 * SD;
WEIGHT = zeros(size(g));
WEIGHT(LIM + 1:end - LIM, LIM + 1:end - LIM) = 1;
[g5, PSF5] = deconvblind(g, INITPSF, NUMIT, DAMPAR, WEIGHT);
imshow(pixeldup(PSF5, 73), [])
```

```matlab
NUMIT = 10;
[g10, PSF10] = deconvblind(g, INITPSF, NUMIT, DAMPAR, WEIGHT);
imshow(pixeldup(PSF10, 73), [])
```
