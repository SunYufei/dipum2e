# 4.2 噪声模型

## 4.2.2 用给定分布产生空间随机噪声

```matlab
r = imnoise2('gaussian', 100000);
subplot(321), hist(r, 50);

r1 = imnoise2('uniform', 100000);
subplot(322), hist(r1, 50);

r2 = imnoise2('lognormal', 100000);
subplot(323), hist(r2, 50);

r3 = imnoise2('rayleigh', 100000);
subplot(324), hist(r3, 50);

r4 = imnoise2('exponential', 100000);
subplot(325), hist(r4, 50);

r5 = imnoise2('erlang', 100000);
subplot(326), hist(r5, 50);
```

## 4.2.3 周期噪声

```matlab
C = [0 64; 0 128; 32 32; 64 0; 128 0; -32 32];
[r, R, S] = imnoise3(512, 512, C);

subplot(121), imshow(S, []);
subplot(122), imshow(r, []);
```

```matlab
C = [0 32; 0 64; 16 16; 32 0; 64 0; -16 16];
[r, R, S] = imnoise3(512, 512, C);

subplot(121), imshow(S, []);
subplot(122), imshow(r, []);
```

```matlab
C = [6 32; -2 2];
[r, R, S] = imnoise3(512, 512, C);
subplot(121), imshow(r, []);

A = [1 5];
[r, R, S] = imnoise3(512, 512, C, A);
subplot(122), imshow(r, []);
```
