## 8.5.1 JPEG

```matlab
f = imread('Tracy.tif');
c1 = im2jpeg(f);
f1 = jpeg2im(c1);
imratio(f, c1)
```

```matlab
compare(f, f1, 3)
```

```matlab
c4 = im2jpeg(f, 4);
f4 = jpeg2im(c4);
imratio(f, c4)
```

```matlab
compare(f, f4, 3)
```

## 8.5.2 JPEG 2000

```matlab
f = imread('Tracy.tif');
c1 = im2jpeg2k(f, 5, [8 8.5]);
f1 = jpeg2k2im(c1);
rms1 = compare(f, f1)
```

```matlab
cr1 = imratio(f, c1)
```

```matlab
c2 = im2jpeg2k(f, 5, [8 7]);
f2 = jpeg2k2im(c2);
rms2 = compare(f, f2)
```

```matlab
cr2 = imratio(f, c2)
```

```matlab
c3 = im2jpeg2k(f, 1, [1 1 1 1]);
f3 = jpeg2k2im(c3);
rms3 = compare(f, f3)
```

```matlab
cr3 = imratio(f, c3)
```
