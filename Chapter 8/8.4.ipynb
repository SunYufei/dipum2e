## Example 8.7

```matlab
f = imread('Brushes.tif');
q = quantize(f, 4, 'igs');
qs = double(q) / 16;
e = mat2lpc(qs);
c = mat2huff(e);
imratio(f, c)
```

```matlab
ne = huff2mat(c);
nqs = lpc2mat(ne);
nq = 16 * nqs;
compare(q, nq)
```

```matlab
compare(f, nq)
```
